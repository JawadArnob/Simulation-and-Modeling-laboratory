```python
import math
import matplotlib.pyplot as plt

S = 20 
# velocity of A,B,C,D
a=3
b=5
c=7
d=2

#x nd y coordinate 
ax=[10]
ay=[0]

bx=[0]
by=[10]

cx=[10]
cy=[10]

dx=[0]
dy=[0]

#shot counter
a_shot=0
b_shot=0
c_shot=0
d_shot=0

for t in range(0,S):
    # print("first is ",t)

    #for A chasing C
    dist_ca = math.sqrt((cx[t]-ax[t])**2 + (cy[t]-ay[t])**2)
    # print("distance between C and A car at {} is {}".format(t,dist_ca))
    
    # shooting condition check
    if (dist_ca < 5):
        print("car C got shot on {}-th minute!!! Target at: ({},{}) & Pursuer at: ({},{})".format(t+1,cx[t],cy[t],ax[t],ay[t]))
        c_shot+=1
    
    cos_theta_ca = (cx[t]-ax[t])/dist_ca
    sin_theta_ca = (cy[t]-ay[t])/dist_ca
    
    ax_new = ax[t] + a*cos_theta_ca
    ay_new = ay[t] + a*sin_theta_ca
    
    ax.append(ax_new)
    ay.append(ay_new)
    
    #for c chasing b
    dist_bc = math.sqrt((bx[t]-cx[t])**2 + (by[t]-cy[t])**2)
    # print("distance between C and A car at {} is {}".format(t,dist_bc))
    
    # shooting condition check
    if (dist_bc < 5):
        print("Car B got shot on {}-th minute!!! Target at: ({},{}) & Pursuer at: ({},{})".format(t+1,bx[t],by[t],cx[t],cy[t]))
        b_shot+=1
    
    cos_theta_bc = (bx[t]-ax[t])/dist_bc
    sin_theta_bc = (by[t]-cy[t])/dist_bc
    
    cx_new = cx[t] + c*cos_theta_bc
    cy_new = cy[t] + c*sin_theta_bc
    
    cx.append(cx_new)
    cy.append(cy_new)

    # for B chasing D
    dist_db = math.sqrt((dx[t]-bx[t])**2 + (dy[t]-by[t])**2)
    # print("distance between C and A car at {} is {}".format(t,dist_db))
    
    # shooting condition check
    if (dist_db < 5):
        print("Car D got shot on {}-th minute!!! Target at: ({},{}) & Pursuer at: ({},{})".format(t+1,dx[t],dy[t],bx[t],by[t]))
        d_shot+=1
    
    cos_theta_db = (dx[t]-bx[t])/dist_db
    sin_theta_db = (dy[t]-by[t])/dist_db
    
    bx_new = bx[t] + b*cos_theta_db
    by_new = by[t] + b*sin_theta_db
    
    bx.append(bx_new)
    by.append(by_new)

    # for D chasing A

    dist_ad = math.sqrt((ax[t]-dx[t])**2 + (ay[t]-dy[t])**2)
    # print("distance between C and A car at {} is {}".format(t,dist_ad))
    
    # shooting condition check
    if (dist_ad < 5):
        print("Car A got shot on {}-th minute!!! Target at: ({},{}) & Pursuer at: ({},{})".format(t+1,ax[t],ay[t],dx[t],dy[t]))
        a_shot+=1
    
    cos_theta_ad = (ax[t]-dx[t])/dist_ad
    sin_theta_ad = (ay[t]-dy[t])/dist_ad
    
    dx_new = bx[t] + d*cos_theta_ad
    dy_new = by[t] + d*sin_theta_ad
    
    dx.append(dx_new)
    dy.append(dy_new)

```

```python
print("for car A ",ax,ay,sep='\n')
print("for car B ",bx,by,sep='\n')
print("for car C ",cx,cy,sep='\n')
print("for car D ",dx,dy,sep='\n')
```

```python
print("Car A got shot {} times".format(a_shot))
print("Car B got shot {} times".format(b_shot))
print("Car C got shot {} times".format(c_shot))
print("Car D got shot {} times".format(d_shot))
```

```python
plt.plot(ax,ay,marker='o',label='Car A')
plt.plot(bx,by,marker='o',label='Car B')
plt.plot(cx,cy,marker='o',label='Car C')
plt.plot(dx,dy,marker='o',label='Car D')
plt.legend()
plt.show()
```
