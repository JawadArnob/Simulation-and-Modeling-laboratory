```python
import numpy as np
np.random.seed(10)

def fifo():
  class SSQ():
    def __init__(self):
        
        self.interarrivals = list(np.abs(np.random.normal(loc=2, scale=1, size=100)))
        self.service_times = list(np.random.uniform(low=3,high=4, size=100))

        print(self.interarrivals)
        print(self.service_times)

        self.clock = 0.0

        self.next_arrival = self.interarrivals.pop(0)
        self.next_departure = float('inf')

        self.num_in_queue = 0
        self.times_of_arrivalqueue = []
        self.service_times_in_queue = []

        self.total_delay = 0.0
        self.num_of_delays = 0.0

        self.server_status = 0 # initially idle
        self.last_event_time = 0.0

        self.avg_delay=0.0

        self.busy_cout=0
        self.total_cout=1 #as 1st one is not considered

        self.utilization=0
        self.clock_list=[]
        self.sum_for_cus=0.0
        self.cus_in_queue=0.0
        


    def arrival(self):
        self.next_arrival = self.next_arrival + self.interarrivals.pop(0)

        if self.server_status == 0: #server idle
            self.server_status = 1
            delay = 0.0
            self.total_delay += delay
            self.num_of_delays += 1
            self.next_departure = self.clock + self.service_times.pop(0)

        else:
            self.num_in_queue += 1
            self.times_of_arrivalqueue.append(self.clock)
            self.service_times_in_queue.append(self.service_times.pop(0))


    def departure(self):
        if self.num_in_queue == 0:
            self.server_status = 0
            next_departure = float('inf')
        else:
            self.num_in_queue -= 1
            arrival = self.times_of_arrivalqueue.pop(0) # FIFO
            delay = self.clock-arrival
            self.total_delay += delay
            self.num_of_delays += 1
            
            self.next_departure = self.clock + self.service_times_in_queue.pop(0)

    def simulate_next_event(self):
        self.clock = min(self.next_arrival, self.next_departure)
        self.last_event_time = self.clock
        self.clock_list.append(self.clock)


        if self.next_arrival<=self.next_departure:
            self.arrival()
            print("arrival at -> " + str(self.clock))
        
        else:
            self.departure()
            print("departure at -> " + str(self.clock))
        self.total_cout+=1
        if self.server_status==1:
            self.busy_cout+=1
        
        print("Server status: ", self.server_status)
        print("Times of arrival in the queue: ", self.times_of_arrivalqueue)
        print("Service times in the queue: ", self.service_times_in_queue)
        print("Total delay: ", self.total_delay)
        print("Next arrival : ", self.next_arrival)
        print("Next departure: ", self.next_departure)
        
        
        


    def start(self):
        while self.num_of_delays < 50:
            self.simulate_next_event()
        self.avg_delay = self.total_delay/self.num_of_delays
        print("avg delay is :",self.avg_delay)

        for i in range(0,len(self.clock_list)-1): #esti customer eq
            self.sum_for_cus+=(self.clock_list[i+1] - self.clock_list[i])*(len(self.service_times_in_queue)) 
        self.cus_in_queue= self.sum_for_cus/self.clock
        print("customers in queue", self.cus_in_queue) 
        
        self.utilization=self.busy_cout/self.total_cout
        print("utilization: ",self.utilization)


  s = SSQ()
  s.start()


def lifo():
  class SSQ():
    def __init__(self):
        
        self.interarrivals = list(np.abs(np.random.normal(loc=2, scale=1, size=100)))
        self.service_times = list(np.random.uniform(low=3,high=4, size=100))

        print(self.interarrivals)
        print(self.service_times)

        self.clock = 0.0

        self.next_arrival = self.interarrivals.pop(0)
        self.next_departure = float('inf')

        self.num_in_queue = 0
        self.times_of_arrivalqueue = []
        self.service_times_in_queue = []

        self.total_delay = 0.0
        self.num_of_delays = 0.0

        self.server_status = 0 # initially idle
        self.last_event_time = 0.0

        self.avg_delay=0.0

        self.busy_cout=0
        self.total_cout=1 #as 1st one is not considered

        self.utilization=0
        self.clock_list=[]
        self.sum_for_cus=0.0
        self.cus_in_queue=0.0


    def arrival(self):
        self.next_arrival = self.next_arrival + self.interarrivals.pop(0)

        if self.server_status == 0: #server idle
            self.server_status = 1
            delay = 0.0
            self.total_delay += delay
            self.num_of_delays += 1
            self.next_departure = self.clock + self.service_times.pop(0)

        else:
            self.num_in_queue += 1
            self.times_of_arrivalqueue.append(self.clock)
            self.service_times_in_queue.append(self.service_times.pop(0))


    def departure(self):
        if self.num_in_queue == 0:
            self.server_status = 0
            next_departure = float('inf')
        else:
            self.num_in_queue -= 1
            arrival = self.times_of_arrivalqueue.pop(-1) # lifo
            delay = self.clock-arrival
            self.total_delay += delay
            self.num_of_delays += 1
            
            self.next_departure = self.clock + self.service_times_in_queue.pop(-1)

    def simulate_next_event(self):
        self.clock = min(self.next_arrival, self.next_departure)

        self.last_event_time = self.clock
        self.clock_list.append(self.clock)


        if self.next_arrival<=self.next_departure:
            self.arrival()
            print("arrival at -> " + str(self.clock))
        
        else:
            self.departure()
            print("departure at -> " + str(self.clock))
        self.total_cout+=1
        if self.server_status==1:
            self.busy_cout+=1
        
        print("Server status: ", self.server_status)
        print("Times of arrival in the queue: ", self.times_of_arrivalqueue)
        print("Service times in the queue: ", self.service_times_in_queue)
        print("Total delay: ", self.total_delay)
        print("Next arrival : ", self.next_arrival)
        print("Next departure: ", self.next_departure)
        
        
        


    def start(self):
        while self.num_of_delays < 50:
            self.simulate_next_event()
        self.avg_delay = self.total_delay/self.num_of_delays
        print("avg delay is :",self.avg_delay)

        for i in range(0,len(self.clock_list)-1): #esti customer eq
            self.sum_for_cus+=(self.clock_list[i+1] - self.clock_list[i])*(len(self.service_times_in_queue)) 
        self.cus_in_queue= self.sum_for_cus/self.clock
        print("customers in queue", self.cus_in_queue) 
        
        self.utilization=self.busy_cout/self.total_cout
        print("utilization: ",self.utilization)


  s = SSQ()
  s.start()

def sjf():
  class SSQ():
    def __init__(self):
        
        self.interarrivals = list(np.abs(np.random.normal(loc=2, scale=1, size=100)))
        self.service_times = list(np.random.uniform(low=3,high=4, size=100))

        print(self.interarrivals)
        print(self.service_times)

        self.clock = 0.0

        self.s=0.0

        self.next_arrival = self.interarrivals.pop(0)
        self.next_departure = float('inf')

        self.num_in_queue = 0
        self.times_of_arrivalqueue = []
        self.service_times_in_queue = []

        self.total_delay = 0.0
        self.num_of_delays = 0.0

        self.server_status = 0 # initially idle
        self.last_event_time = 0.0

        self.avg_delay=0.0

        self.busy_cout=0
        self.total_cout=1 #as 1st one is not considered

        self.utilization=0
        self.clock_list=[]
        self.sum_for_cus=0.0
        self.cus_in_queue=0.0



    def arrival(self):
        self.next_arrival = self.next_arrival + self.interarrivals.pop(0)

        if self.server_status == 0: #server idle
            self.server_status = 1
            delay = 0.0
            self.total_delay += delay
            self.num_of_delays += 1
            self.next_departure = self.clock + self.service_times.pop(0)

        else:
            self.num_in_queue += 1
            self.times_of_arrivalqueue.append(self.clock)
            self.service_times_in_queue.append(self.service_times.pop(0))


    def departure(self):
        self.s=self.service_times_in_queue.index(min(self.service_times_in_queue))
        
        if self.num_in_queue == 0:
            self.server_status = 0
            next_departure = float('inf')
        else:
            self.num_in_queue -= 1
            arrival = self.times_of_arrivalqueue.pop(self.s) # sjf
            delay = self.clock-arrival
            self.total_delay += delay
            self.num_of_delays += 1
            
            self.next_departure = self.clock + self.service_times_in_queue.pop(self.s)

    def simulate_next_event(self):
        self.clock = min(self.next_arrival, self.next_departure)
        self.last_event_time = self.clock
        self.clock_list.append(self.clock)

        if self.next_arrival<=self.next_departure:
            self.arrival()
            print("arrival at -> " + str(self.clock))
        
        else:
            self.departure()
            print("departure at -> " + str(self.clock))
        self.total_cout+=1
        if self.server_status==1:
            self.busy_cout+=1
        print("clock :",self.clock)
        print("Server status: ", self.server_status)
        print("Times of arrival in the queue: ", self.times_of_arrivalqueue)
        print("Service times in the queue: ", self.service_times_in_queue)
        print("Total delay: ", self.total_delay)
        print("Next arrival : ", self.next_arrival)
        print("Next departure: ", self.next_departure)
        
        
        


    def start(self):
        while self.num_of_delays < 50:
            self.simulate_next_event()
        self.avg_delay = self.total_delay/self.num_of_delays
        print("avg delay is :",self.avg_delay)

        for i in range(0,len(self.clock_list)-1): #esti customer eq
            self.sum_for_cus+=(self.clock_list[i+1] - self.clock_list[i])*(len(self.service_times_in_queue)) 
        self.cus_in_queue= self.sum_for_cus/self.clock
        print("customers in queue", self.cus_in_queue) 

        self.utilization=self.busy_cout/self.total_cout
        print("utilization: ",self.utilization)


  s = SSQ()
  s.start()
x=int(input("1.fifo 2.Lifo 3.Sjf: "))

if x==1:
  fifo()
elif x==2:
  lifo()
elif x==3:
  sjf()
```
