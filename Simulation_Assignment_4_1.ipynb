```python
CarA = [(10, 0)]
CarB = [(0, 10)]
CarC = [(10, 10)]
CarD = [(0, 0)]
Shoot_dist = {'AC' : [],'CB' : [],'BD' : [],'DA' : []}
```

```python
def new_cordiante(a,b,dist,S) :
    if dist == 0 :
        return a
    else :
        cos_theta = (b[0]-a[0])/dist
        sin_theta = (b[1]-a[1])/dist
        ax_new = a[0] + S*cos_theta
        ay_new = a[1] + S*sin_theta
        return (ax_new, ay_new)
```

```python
import math

SA,SB,SC,SD = (3,5,7,2)

for t in range(0,20):
    dist_AC = math.sqrt((CarA[t][0]-CarC[t][0])**2 + (CarA[t][1]-CarC[t][1])**2)
    dist_CB = math.sqrt((CarB[t][0]-CarC[t][0])**2 + (CarB[t][1]-CarC[t][1])**2)
    dist_BD = math.sqrt((CarB[t][0]-CarD[t][0])**2 + (CarB[t][1]-CarD[t][1])**2)
    dist_DA = math.sqrt((CarA[t][0]-CarD[t][0])**2 + (CarA[t][1]-CarD[t][1])**2)
    if dist_AC <= 5 :
        Shoot_dist['AC'].append(t+1)
    if dist_CB <= 5 :
        Shoot_dist['CB'].append(t+1)
    if dist_BD <= 5 :
        Shoot_dist['BD'].append(t+1)
    if dist_DA <= 5:
        Shoot_dist['DA'].append(t+1)
      
    CarA_new = new_cordiante(CarA[t],CarC[t],dist_AC,SA)
    CarC_new = new_cordiante(CarC[t],CarB[t],dist_CB,SC)
    CarB_new = new_cordiante(CarB[t],CarD[t],dist_BD,SB)
    CarD_new = new_cordiante(CarD[t],CarA[t],dist_DA,SD)
    
    CarA.append(CarA_new)
    CarB.append(CarB_new)
    CarC.append(CarC_new)
    CarD.append(CarD_new)
    
```

```python
import matplotlib.pyplot as plt
```

```python
#x and y coordinate
print(CarA,CarB,CarC,CarD,sep='\n')
# print(CarB)
# print(CarC)
# print(CarD)
```

```python
#print hit count and time of getting shot
print('Car A got shoot at {} seconds and in total  {} times '.format(Shoot_dist['DA'],len(Shoot_dist['DA'])))
print('Car B got shoot at {} seconds and in total  {} times '.format(Shoot_dist['CB'],len(Shoot_dist['CB'])))
print('Car C got shoot at {} seconds and in total  {} times '.format(Shoot_dist['AC'],len(Shoot_dist['AC'])))
print('Car D got shoot at {} seconds and in total  {} times '.format(Shoot_dist['BD'],len(Shoot_dist['BD'])))
```

```python
#print figure
plt.plot(*zip(*CarA) ,marker='o',label='A')
plt.plot(*zip(*CarB) ,marker='o',label='B')
plt.plot(*zip(*CarC) ,marker='o',label='C')
plt.plot(*zip(*CarD) ,marker='o',label='D')
plt.legend()
plt.show()
```
